```python
import numpy as np
import matplotlib.pyplot as plt

#parameter values
n_flips = 10
n_experiments = 1000

#Here we simulate 100 experiments of 10 coin flips each
results = np.random.binomial(n=n_flips, p=0.5, size=n_experiments)

#Now plot the distribution of results
plt.hist(results, bins=np.arange(-0.5, n_flips+1, 1), edgecolor='black')
plt.xlabel('Number of Heads')
plt.ylabel('Frequency')
plt.title('Distribution of Heads in 10 Coin Flips (1000 Experiments)')
plt.show()

#Probability of eactly 5 heads
prob_5_heads = np.mean(results == 5)
print("Probability of exactly 5 heads:" + str(prob_5_heads))
```
